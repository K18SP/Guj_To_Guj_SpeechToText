```python
from moviepy.editor import AudioFileClip  # Import AudioFileClip

def extract_audio(audio_path, output_path):
    """Extracts audio from an audio-only MP4 file and saves it as a WAV file."""
    audio = AudioFileClip(audio_path)  # Load the audio-only MP4 as an AudioFileClip
    audio.write_audiofile(output_path)  # Save the audio to a WAV file

# Example usage
audio_path = "/content/sample.02.40_9851572d.dat.mp4"  # Your audio-only MP4 file
output_path = "output_audio.wav"
extract_audio(audio_path, output_path)

# Rest of your code (audio_to_text, summarize_text, video_to_summary) remains the same
```

```python
import speech_recognition as sr

def audio_to_text(audio_path):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio = recognizer.record(source)
    try:
        text = recognizer.recognize_google(audio, language="gu-IN")  # For Gujarati
        # text = recognizer.recognize_google(audio, language="en-US")  # For English
        return text
    except sr.UnknownValueError:
        return "Google Speech Recognition could not understand audio"
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"

# Example usage
text = audio_to_text(audio_path)
print("Transcribed Text:", text)
```

```python
from moviepy.editor import AudioFileClip
import speech_recognition as sr

def audio_mp4_to_text(audio_mp4_path, language="gu-IN"):
    """
    Extracts audio from an audio-only MP4 file, converts it to WAV,
    and then transcribes the audio to text using Google Speech Recognition.
    """
    try:
        # 1. Extract audio to a temporary WAV file
        audio = AudioFileClip(audio_mp4_path)
        temp_wav_path = "temp_audio.wav"  # Temporary file for extracted audio
        audio.write_audiofile(temp_wav_path)

        # 2. Transcribe the audio to text
        recognizer = sr.Recognizer()
        with sr.AudioFile(temp_wav_path) as source:
            audio_data = recognizer.record(source)
        text = recognizer.recognize_google(audio_data, language=language)

        return text

    except sr.UnknownValueError:
        return "Google Speech Recognition could not understand audio"
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"
    except Exception as e:  # Catch other potential errors (e.g., file not found)
        return f"An error occurred: {e}"
    finally:
        # Clean up the temporary WAV file (if it was created)
        import os
        if os.path.exists(temp_wav_path):
            os.remove(temp_wav_path)

# Example usage
audio_mp4_path = "/content/sample.02.40_9851572d.dat.mp4"
transcribed_text = audio_mp4_to_text(audio_mp4_path, language="gu-IN")
print("Transcribed Text:", transcribed_text)
```
